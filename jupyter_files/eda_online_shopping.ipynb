# Sales Analysis - 2019

This section of notebook provides a structured exploration of sales performance, summarizing key indicators and trends to support data-driven decision-making.  

## Key Metrics
- **Sum of Gross Sales** -> overall revenue before deductions.  
- **Sum of Net Sales** -> revenue after discounts and charges.  
- **Total Orders** -> number of completed transactions.  
- **Unique Customers** -> customer base contributing to sales.  
- **Average Order Value (AOV)** ->revenue per order.  
- **Average Discount %** -> average promotional discount applied.  

## Revenue Analysis
- **Revenue by Gender** -> differences in buying behavior.  
- **Revenue by Channel** -> comparison of online vs. offline sales based on all-time spend.  
- **Revenue by Location** -> geographic distribution of revenue.  
- **Revenue by Discount Range** -> impact of discounts on revenue.  
- **Revenue by Tenure Group** -> spending behavior across customer tenure.  

## Growth and Trends
- **Monthly Revenue** -> recurring performance across time.  
- **Monthly Growth** -> rate of increase or decrease in sales.  

## Top Performers
- **Top 10 Product Categories** -> highest revenue-generating categories.  
- **Top 10 Customers** -> high-value customers based on profitability.

```python
#Create Gross Sales column
#Gross sales = Quantity * Average Price 
df["Gross_Sales"] = df["Quantity"]*df["Avg_Price"]

#Create Net Sales column
#Net Sales = Gross Sales * (1 - Discount/100) + Delivery Charges + GST
df["Net_Sales"] = df["Gross_Sales"] * (1 - df["Discount_pct"]/100) + df['Delivery_Charges'] + df["GST"]
```

```python
df.info()
```

```python
#Rename CustomerID to Customer_ID for consistency
df.rename(columns={'CustomerID':'Customer_ID'},inplace=True)
```

```python
df.info()
```

```python
print("=== KEY PERFORMANCE INDICATORS (KPIs) ===")

#Total Gross Sales
total_gs = df['Gross_Sales'].sum()
print(f"Total Gross Sales: ${total_gs:,.2f}")

#Total Net Sales
total_ns = df['Net_Sales'].sum()
print(f"Total Net Sales: ${total_ns:,.2f}")

#Total Customers
customers = df['Customer_ID'].nunique()
print(f"Total Customers: {customers:,.0f}")

#Total Orders
orders = df['Transaction_ID'].nunique()
print(f"Total Orders: {orders:,.0f}")

#Average Value Order
aov = total_ns/orders
print(f"AOV: ${aov:,.2f}")

#Average Discount %
avg_disc = df['Discount_pct'].mean()
print(f"Average Discount: {avg_disc:,.1f}%")

#Revenue by Gender
rev_gender = df.groupby('Gender', observed=False)['Net_Sales'].sum().apply(lambda x: f"${x:,.2f}") #Added observed=False to include all possible categories in the output


#Revenue by Channel based on all-time spend
rev_channel = df.groupby("Customer_ID")[["Online_Spend","Offline_Spend"]].max().sum().apply(lambda x: f"${x:,.2f}")
#rev_channel = customer_spend.sum().apply(lambda x: f"${x:,.2f}")

#Revenue by Location
rev_loc = df.groupby('Location')['Net_Sales'].sum()

#Monthly Revenue & Growth 
df['Order_Month'] = df['Transaction_Date'].dt.to_period('M')
monthly_rev = df.groupby('Order_Month')['Net_Sales'].sum()
monthly_growth = monthly_rev.pct_change()*100

#Top 10 Product Categories 
top_prod = df.groupby('Product_Category')['Net_Sales'].sum().sort_values(ascending=False).head(10)

#Top 10 Customers
top_cst = df.groupby('Customer_ID')['Net_Sales'].sum().sort_values(ascending=False).head(10)

#Revenue by Discount Range/Band
#--Create disount range labels
discount = [0, 10, 30, 100]
disc_range = ['0-10%','11-30%', '31%+']
df['Disc_Range'] = pd.cut(df['Discount_pct'], bins=discount, labels=disc_range, include_lowest=True)

rev_disc = df.groupby('Disc_Range', observed=True)['Net_Sales'].sum()

#Revenue by Tenure Group
#--Create tenure group labels
tenure = [0, 12, 36, 100]
ten_range = ['0-12', '13-36', '37+']
df['Tenure_Group'] = pd.cut(df['Tenure_Months'], bins=tenure, labels=ten_range, include_lowest=True)

rev_ten = df.groupby('Tenure_Group', observed=True)['Net_Sales'].sum()
```
